# Feature Engineering

## Categorical variable

**Splitting the email id's based on the data as professional, Personal, Junk**

```python
# define function to categorize domains
def categorize_domain(domain):
    personal_domains = ['gmail.com', 'yahoo.com', 'protonmail.com']
    junk_domains = ['me.com', 'aol.com', 'hotmail.com', 'live.com', 'msn.com', 'passport.com']
    if domain in personal_domains:
        return 'PERSONAL'
    elif domain in junk_domains:
        return 'JUNK'
    else: 
        return 'PROFESSIONAL'

# extract domain name from email addresses
data['SS_EMAIL_CATEGORY'] = data['EMAIL'].apply(lambda x: x.split('@')[1])

# apply domain categorization function
data['SS_EMAIL_CATEGORY'] = data['SS_EMAIL_CATEGORY'].apply(categorize_domain)

# print result
data.head()
```

**Converting the Categorical variable to Numerical Variable**

```python
encoded_data = pd.get_dummies(data['SS_EMAIL_CATEGORY'], prefix='SS_EMAIL_CATEGORY',drop_first=True)

# Add the one-hot encoded columns to the original DataFrame
data = pd.concat([data, encoded_data], axis=1)
data.head(2)
```

## Numerical Data

```python
#Revenue per total meal ordered.
data['FE_REVENUE_PER_MEAL'] = data['REVENUE']/data['TOTAL_MEALS_ORDERED']
#Revenue per login.
data['FE_REVENUE_PER_LOGIN']= data['REVENUE']/(data['PC_LOGINS']+data['MOBILE_LOGINS'])
#Unique meal purch per total meals ordered
data['FE_UNIQUE_MEAL_ORDERED_PER_TOTAL_ORDERED']=data['UNIQUE_MEALS_PURCH']/data['TOTAL_MEALS_ORDERED']
#Total photo viewed per avg time per site visit
data['FE_PHOTOS_VIEWED_PER_AVG_TIME_SPENT']=data['TOTAL_PHOTOS_VIEWED']/data['AVG_TIME_PER_SITE_VISIT']
#Avg num of meal ordered per customer per total meal ordered
data['FE_AVG_NUM_OF_MEAL_ORDERED_PER_CUSTOMER']=data['AVG_NUM_OF_MEAL_ORDERED_PER_CUSTOMER']/data['TOTAL_MEALS_ORDERED']
#generating feature as ratio of service requests per meal ordered
data['FE_CONTACTS_W_CUSTOMER_SERVICE'] =  data['TOTAL_MEALS_ORDERED'] / data['CONTACTS_W_CUSTOMER_SERVICE']
#cancellation after noon per meal ordered
data['FE_CANCELLATION_AFTER_NOON_PMO'] = data['CANCELLATIONS_AFTER_NOON'] / data['TOTAL_MEALS_ORDERED']

```
